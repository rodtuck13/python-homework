```python
from pathlib import Path
import pandas as pd
import csv
```

```python
menu_data = Path ("C:/Users/Rodtwo/Documents/FinTech_BootCamp/python-homework/PyRamen/menu_data.csv")
```

```python
menu = []
with open (menu_data, 'r') as menudata:
    menureader = csv.reader(menudata, delimiter = ',')
    header = next(menureader)
        
    for row in menureader:
        hold = [row[0],row[1],row[2],float(row[3]),float(row[4])]
        menu.append(hold)
```

```python
sales_data = Path ("C:/Users/Rodtwo/Documents/FinTech_BootCamp/python-homework/PyRamen/sales_data.csv")
```

```python
sales = []
with open (sales_data, 'r') as salesdata:
    salesreader = csv.reader(salesdata, delimiter = ',')
    header = next(salesreader)
        
    for row in salesreader:
        hold = [row[0],row[1],row[2],int(row[3]),row[4]]
        sales.append(hold)
```

```python
sales[:10]
```

```python
new_dict = {}

for n in range(len(sales)):
    menu_item = sales[n][4]
    quantity = sales[n][3]
    counter = 0
    if menu_item not in new_dict:
        new_dict[menu_item] = {
            "01-count" : 0,
            "02-revenue" : 0,
            "03-cogs" : 0,
            "04-profit" : 0
        }
    for s in range(len(menu)):
        item = menu[s][0]
        price = menu[s][3]
        cost = menu[s][4]
        profit = price - cost
        if menu_item == item:
            new_dict[menu_item]["01-count"] += quantity
            new_dict[menu_item]["02-revenue"] += quantity * price
            new_dict[menu_item]["03-cogs"] += quantity * cost
            new_dict[menu_item]["04-profit"] += quantity * profit
        elif menu_item != item:
            counter += 1
            if counter == len(menu):
                print(f"{menu_item} does not equal {item}! NO MATCH!")
          
            
new_dict
```

```python
with open('report.txt','w') as report:
    print(new_dict, file = report)
```
